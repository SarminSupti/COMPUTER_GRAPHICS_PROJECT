Task 2: Building a House in Rainfall
Draw a house with a raindrop using the base primitives: points, lines, or triangles. You can use ONLY GL_POINTS, GL_LINES, or GL_TRIANGLES for designing this house. A diagram has been provided as an example. You can modify the house design to your liking. The rain drops should be animated to fall from top to bottom.
It has been raining unwantedly for the last few days, so let’s control its direction by designing a key that will change the direction of the rain when clicked (slightly bending the rainfall). Design this functionality such that the left arrow will gradually bend the rain to the left and the right arrow will gradually bend the rain to the right.
Formulate two more keys(assign whatever key you like); pressing one will gradually change the skin colour from dark to light simulating night to day, and the other will change it from light to dark simulating day to night . You must also consider the rain and the house visibility in different background colours.

**You can’t use glutTimerFunc() at all


```python
#Task-2 solution

from OpenGL.GL import *
from OpenGL.GLUT import *
from OpenGL.GLU import *
import random

rain_drops = []
rain_speed = 3
rain_direction = 0
rain_direction_limit = 3.0
D_N_color = [0, 0, 0]

def draw_house(x, y):
    glPointSize(5)

    glBegin(GL_LINES)
    glVertex2f(0,300)
    glVertex2f(500,300)
    glEnd()

    glBegin(GL_LINES)
    glColor3f(0.5, 0.4, 0.12)
    for i in range(0,301):
        glVertex2f(0,i)
        glVertex2f(500,i)
    glEnd()

    glBegin(GL_TRIANGLES)    #roof
    glColor3f(0.47, 0.25, 0.01)
    glVertex2f(250,350)
    glVertex2f(100,250)
    glVertex2f(400,250)
    glEnd()

    glBegin(GL_TRIANGLES)   #House_wall
    glColor3f(0.72, 0.71, 0.09)
    glVertex2f(385,140)
    glVertex2f(115,250)
    glVertex2f(385,250)
    glVertex2f(385,140)
    glVertex2f(115,250)
    glVertex2f(115,140)
    glEnd()


    glBegin(GL_TRIANGLES)     #door
    glColor3f(0.459, 0.455, 0.427)
    glVertex2f(200,140)
    glVertex2f(200,200)
    glVertex2f(300,140)
    glVertex2f(200,200)
    glVertex2f(300,200)
    glVertex2f(300,140)
    glEnd()

    glBegin(GL_TRIANGLES)     #window
    glColor3f(0.459, 0.455, 0.427)

    glVertex2f(140,230)
    glVertex2f(170, 230)
    glVertex2f(170,190)
    glVertex2f(140,230)
    glVertex2f(140,190)
    glVertex2f(170,190)

    glVertex2f(330,230)
    glVertex2f(360, 230)
    glVertex2f(360,190)
    glVertex2f(330,230)
    glVertex2f(330,190)
    glVertex2f(360,190)
    glEnd()

    glBegin(GL_LINES)
    glColor3f(0, 0, 0)
    glVertex2f(140,210)
    glVertex2f(170, 210)

    glVertex2f(155,230)
    glVertex2f(155, 190)

    glVertex2f(345,230)
    glVertex2f(345, 190)

    glVertex2f(330,210)
    glVertex2f(360, 210)

    glEnd()

def idle():
    update_rain()
    glutPostRedisplay()

def rain_fall():
    global rain_drops
    for i in range(150):
        x= random.randint(0,500)
        y= random.randint(0,500)
        rain_drops.append([x,y])

def draw_rain():

    glColor3f(0.5, 0.5, 1.0)
    glBegin(GL_LINES)

    for d in rain_drops:
        glVertex2f(d[0], d[1])
        glVertex2f(d[0]+rain_direction, d[1] - 10)
    glEnd()

def update_rain():
    global rain_drops
    for d in rain_drops:
        d[0] += rain_direction *0.5
        d[1] -= rain_speed
        if d[1] < 0 or d[0] < 0 or d[0] > 500:
            d[0] = random.randint(0, 500)
            d[1] = 500

def special_keys( key, x, y):
    global rain_direction

    if key == GLUT_KEY_LEFT:
        if rain_direction > -rain_direction_limit:
            rain_direction -= 0.2
    elif key == GLUT_KEY_RIGHT:
        if rain_direction < rain_direction_limit:
            rain_direction += 0.2

def keyboard(key, x, y):

    global D_N_color
    if key == b'd':

        if D_N_color[0] < 1:
            D_N_color[0] += 0.1
        if D_N_color[1] < 1:
            D_N_color[1] += 0.1
        if D_N_color[2] < 1:
            D_N_color[2] += 0.1

    elif key == b'n':

        if D_N_color[0] > 0:
            D_N_color[0] -= 0.1
        if D_N_color[1] > 0:
            D_N_color[1] -= 0.1
        if D_N_color[2] > 0:
            D_N_color[2] -= 0.1
    glutPostRedisplay()


def iterate():
    glViewport(0, 0, 500, 500)
    glMatrixMode(GL_PROJECTION)
    glLoadIdentity()
    glOrtho(0.0, 500, 0.0, 500, 0.0, 1.0)
    glMatrixMode (GL_MODELVIEW)
    glLoadIdentity()

def showScreen():
    global D_N_color
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
    glClearColor(*D_N_color, 1.0)
    glLoadIdentity()
    iterate()
    glColor3f(1.0, 1.0, 0.0)
    draw_house(250, 250)
    draw_rain()
    glutSwapBuffers()


glutInit()
glutInitDisplayMode(GLUT_RGBA)
glutInitWindowSize(500, 500) #window size
glutInitWindowPosition(0, 0)
wind = glutCreateWindow(b"House with Rain Direction and Day-Night") #window name

rain_fall()
glutDisplayFunc(showScreen)
glutIdleFunc(idle)
glutSpecialFunc(special_keys)
glutKeyboardFunc(keyboard)

glutMainLoop()
```
